```python
import urllib3
import certifi
import csv
from bs4 import BeautifulSoup
```

```python
url = 'https://www.1001tracklists.com/tracklist/2gd5uhbk/solid-stone-refresh-radio-188-2018-02-22.html'
http = urllib3.PoolManager(
    cert_reqs='CERT_REQUIRED',
    ca_certs=certifi.where())

response = http.request('GET', url)
soup = BeautifulSoup(response.data, "lxml")
```

```python
soup
```

```python
name_box = soup.find_all('div', attrs={'class', 'tlToogleData'})
name_list = []
for names in name_box:
    name_list.append(names.text.strip())
name_box    
```

```python
final = []
for name in name_list:
    head, sep, tail = name.partition('  [')
    final.append(head)
```

```python
final = [f.replace('\xa0', ' ') for f in final]
```

```python
file = open('solid_stone.txt', 'w')
for val in final:
    var1 = val.split(',');
    file.write('%s\n' % var1)
file.close()
```
